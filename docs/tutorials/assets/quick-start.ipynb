# Import libraries

```python
import mage_ai
```

# Load data

```python
from mage_ai.server.sample_datasets import list_dataset_names, load_dataset


dataset_names = list_dataset_names()
print(dataset_names)
df = load_dataset(dataset_names[0])
```

```python
df
```

# Connect data to app

```python
mage_ai.connect_df(df, 'titanic survival dataset')
```

# Run app locally

```python
mage_ai.kill() # Kill any previously running instance
mage_ai.launch()
```

# Clean data using pipeline created from app

```python
mage_ai.clean(df, pipeline_uuid=1)
```
